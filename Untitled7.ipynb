```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
```

```python
X = df.drop('Class', axis=1)
y = df['Class']

print(y.value_counts())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,          # VERY IMPORTANT for imbalanced data
    random_state=42
)
```

```python
print("Train Shape:", X_train.shape)
print("Test Shape:", X_test.shape)
```

```python
#Feature Scaling
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Handle Class Imbalance (SMOTE)
smote = SMOTE(random_state=42)

X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

print("after SMOTE:")
print(y_train_resampled.value_counts())
```

```python
print("Final Training Shape:", X_train_resampled.shape)
print("Final Testing Shape:", X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
X_train_df = pd.DataFrame(X_train_resampled, columns=X.columns)
X_test_df = pd.DataFrame(X_test, columns=X.columns)
```

```python
# Create new feature: Transaction Amount Category
df['Amount_Category'] = pd.qcut(df['Amount'], q=4, labels=False)

# Create new feature: High Risk Transaction Flag
df['High_Amount'] = (df['Amount'] > df['Amount'].quantile(0.95)).astype(int)
```

```python
# Feature Selection
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_df, y_train_resampled)

importances = pd.Series(rf.feature_importances_, index=X_train_df.columns)

top_features = importances.nlargest(15).index

print("Top Selected Features:")
print(top_features)
```

```python
X_train_selected = X_train_df[top_features]
X_test_selected = X_test_df[top_features]
```

```python
X_train_final = X_train_selected
X_test_final = X_test_selected

print("Final Feature Shape:")
print(X_train_final.shape, X_test_final.shape)
```

## Model Building (Machine Learning)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, roc_auc_score
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM": SVC(probability=True)
}
```

```python
results = {}

for name, model in models.items():
    print(f"\n{name}")

    # Train
    model.fit(X_train_final, y_train_resampled)

    # Predict
    y_pred = model.predict(X_test_final)
    y_prob = model.predict_proba(X_test_final)[:,1]

    # Metrics
    print(classification_report(y_test, y_pred))

    roc = roc_auc_score(y_test, y_prob)
    print("ROC-AUC:", roc)

    results[name] = roc
```

```python
import pandas as pd

results_df = pd.DataFrame(list(results.items()), columns=["Model", "ROC-AUC"])
results_df = results_df.sort_values(by="ROC-AUC", ascending=False)

print("\nModel Comparison:")
print(results_df)
```

```python
best_model_name = results_df.iloc[0]['Model']
print("Best Model:", best_model_name)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5]
}

grid = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)

grid.fit(X_train_final, y_train_resampled)

print("\nBest Parameters:", grid.best_params_)
print("Best Score:", grid.best_score_)

# Best model
best_rf = grid.best_estimator_
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier

param_dist = {
    'n_estimators': [50, 100, 150],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5, 10]
}

random_search = RandomizedSearchCV(
    RandomForestClassifier(random_state=42),
    param_distributions=param_dist,
    n_iter=5,              # only 5 combinations instead of all
    cv=3,                  # reduce folds
    scoring='roc_auc',
    n_jobs=-1,
    random_state=42
)

random_search.fit(X_train_final, y_train_resampled)

print("Best Params:", random_search.best_params_)
print("Best Score:", random_search.best_score_)

best_rf = random_search.best_estimator_
```

```python
y_pred = best_rf.predict(X_test_final)
y_prob = best_rf.predict_proba(X_test_final)[:,1]

print("\nTuned Model Performance:")
print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```

## ANN Model Development

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
ann = Sequential()
ann.add(Dense(units=32, activation='relu', input_dim=X_train_final.shape[1]))
ann.add(Dropout(0.3))
ann.add(Dense(units=16, activation='relu'))
ann.add(Dropout(0.2))
ann.add(Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
history = ann.fit(
    X_train_final,
    y_train_resampled,
    epochs=10,
    batch_size=64,
    validation_split=0.2
)
```

```python
y_pred_ann = (ann.predict(X_test_final) > 0.5).astype(int)

from sklearn.metrics import classification_report, roc_auc_score

print("\nANN Performance:")
print(classification_report(y_test, y_pred_ann))
print("ROC-AUC:", roc_auc_score(y_test, ann.predict(X_test_final)))
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

roc_scores = []

for train_idx, val_idx in skf.split(X_train_final, y_train_resampled):

    X_tr = X_train_final.iloc[train_idx]
    X_val = X_train_final.iloc[val_idx]
    y_tr = y_train_resampled.iloc[train_idx]
    y_val = y_train_resampled.iloc[val_idx]

    ann_cv = Sequential([
        Dense(32, activation='relu', input_dim=X_tr.shape[1]),
        Dropout(0.3),
        Dense(16, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])

    ann_cv.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='binary_crossentropy'
    )

    ann_cv.fit(X_tr, y_tr, epochs=5, batch_size=64, verbose=0)

    y_prob = ann_cv.predict(X_val)
    score = roc_auc_score(y_val, y_prob)

    roc_scores.append(score)

print("Cross-validated ROC-AUC:", sum(roc_scores)/len(roc_scores))
```
